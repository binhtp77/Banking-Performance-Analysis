```python
#Convert credit_limits: str to int
df_cards['credit_limit'] = pd.to_numeric(
    df_cards['credit_limit'].str.replace(r'[^0-9]', '', regex=True),
    errors='coerce'
)
```

```python
#Convert credit_limits: str to date
df_cards['expires'] = pd.to_datetime(
    df_cards['expires'],
    format='%m/%Y'
) + pd.offsets.MonthEnd(0)
```

```python
#Convert credit_limits: str to date
df_cards['acct_open_date'] = pd.to_datetime(
    df_cards['acct_open_date'],
    format='%m/%Y'
)
```

```python
df_cards['has_chip'] = df_cards['has_chip'].replace({
    'YES' : "Yes",
    'NO' : "No"
})
```

```python
#Replace values in df_cards['has_chip']
df_cards['has_chip'] = df_cards['has_chip'].replace('No', False)
```

```python
engine = create_engine(
    "mysql+pymysql://root:insertyourpassword@localhost:3306/dataset_project_banking"
)

def load_to_mysql(df, table_name, if_exists="append"):
    df = df.where(pd.notnull(df), None)
    df.columns = df.columns.str.replace(" ", "_")

    df.to_sql(
        name=table_name,
        con=engine,
        if_exists=if_exists,
        index=False,
        chunksize=1000
    )

    print(f"✅ Loaded vào bảng: {table_name}")
```

```python
load_to_mysql(df_transactions, "fact_transactions")
load_to_mysql(df_users, "dim_users")
load_to_mysql(df_mcc, "dim_mcc")
```
